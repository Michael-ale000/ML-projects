```python
#independent variable('height', 'age', 'appearance', 'goals', 'assists', 'yellow cards', 'second yellow cards', 'red cards',
#'goals conceded', 'clean sheets', 'minutes played', 'days_injured', 'games_injured','highest price' 'position_encoded', 'winger')
X=ds.iloc[:,np.r_[4:18,19:22]].values  
```

```python
#independent variable('current price')
y=ds.iloc[:,18]
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
```

```python
std
```

```python
X_scaled=std.fit_transform(X)
```

```python
X_scaled
```

# splitting data


```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=0)
```

# applying ml algorithms

# 1)  linear regression 

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
```

```python
lr
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error

# Assuming y_pred and y_test are your predicted and true values, respectively
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)
```

```python
#R-square accuracy measure
#R-squared measures the proportion of the variance in the dependent variable that is predictable from the independent variables.
# It ranges from 0 to 1, where 1 indicates a perfect fit.
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print('R-squared:', r2)
```

```python
plt.figure(figsize=(12,6))
plt.plot(y_test,'blue',label='Original value')
plt.plot(y_pred,'red',label='Predicted value')
plt.legend()
plt.show()
```

```python
# we get average accuracy using linear regression
```

# applying simple neural network


```python
#importing necessary libraries for NN
import tensorflow as tf
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)), #first layer with inputs and 64 nodes
    tf.keras.layers.Dense(32, activation='relu'),  # second layer with 32 nodes
    tf.keras.layers.Dense(1)  # Output layer with one neuron for regression
])
```

```python
#compiling the model
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.fit(X_train,y_train,epochs=50,batch_size=32,validation_data=(X_test,y_test))
```

```python
model.save("Football Players Transfer Fee prediction_model.h5")
```

```python
loss = model.evaluate(X_test, y_test)
print(f'Mean Squared Error on Test Set: {loss}')
```
